```python
import numpy as np
```

Adaptive CSR

```python
m = 256 #H
n = 96 #W
for sparsity_percent in range(1,100):
    sparsity = sparsity_percent/100
    k = np.ceil(m*n*sparsity) #nonzero numbers
    offset_storage = (m + 1) * np.ceil(np.log2(k))
    col_idx_storage = k * np.ceil(np.log2(n))
    value_storage = k
    compress_ratio = (offset_storage + col_idx_storage + value_storage) / (m*n)
    print(f'm,n,sparsity:{m,n,1-sparsity} offset_storage:{offset_storage/8}B {int(offset_storage/m)}b col_idx_storage:{col_idx_storage/8}B {int(col_idx_storage/k)}b compress_storage:{(offset_storage + col_idx_storage + value_storage)/8} Bytes, dense_storage:{m*n/8} Bytes, compress_ratio:{1-compress_ratio}')
```

COO format

```python
m = 128
n = 128
for sparsity_percent in range(1,100):
    sparsity = sparsity_percent/100
    k = np.ceil(m*n*sparsity)
    coo_storage = k * np.ceil(np.log2(m)) * 2
    value_storage = k
    compress_ratio = (coo_storage + value_storage) / (m*n)
    print(f'm,n,sparsity:{m,n,1-sparsity} coo_storage:{coo_storage/8}B {int(coo_storage/k)}b compress_storage:{(coo_storage + value_storage)/8} Bytes, dense_storage:{m*n/8} Bytes, compress_ratio:{1-compress_ratio}')
```

Place timestep in row dimension (not efficient)

```python
m = 64*8
n = 64
for sparsity_percent in range(1,100):
    sparsity = sparsity_percent/100
    k = np.ceil(m*n*sparsity)
    offset_storage = m * np.ceil(np.log2(k))
    col_idx_storage = k * np.ceil(np.log2(n))
    value_storage = k
    compress_ratio = (offset_storage + col_idx_storage + value_storage) / (m*n)
    print(f'm,n,sparsity:{m,n,sparsity} {offset_storage, col_idx_storage} compress_storage:{(offset_storage + col_idx_storage + value_storage)/8} Bytes, dense_storage:{m*n/8} Bytes, compress_ratio:{1-compress_ratio}')
```

Fix bandwidth implementation

```python
m = 32
n = 32
offset_precion = 16
col_idx_precision = 16
for sparsity_percent in range(1,100):
    sparsity = sparsity_percent/100
    k = np.ceil(m*n*sparsity)
    offset_storage = m * offset_precion
    col_idx_storage = k * col_idx_precision
    value_storage = k
    compress_ratio = (offset_storage + col_idx_storage + value_storage) / (m*n)
    print(f'm,n,sparsity:{m,n,1-sparsity} offset_storage:{offset_storage}B col_idx_storage:{col_idx_storage}B compress_storage:{(offset_storage + col_idx_storage + value_storage)/8} Bytes, dense_storage:{m*n/8} Bytes, compress_ratio:{1-compress_ratio}')
```

Tiling

```python
M = 128
N = 128
tile = 4
for sparsity_percent in range(1,100):
    sparsity = sparsity_percent/100
    m = M/tile
    n = N/tile    
    k = np.ceil(m*n*sparsity)
    offset_precision = int(np.ceil(np.log2(k)))
    col_idx_precision = int(np.ceil(np.log2(n)))
    offset_storage = m * offset_precision *tile*tile
    col_idx_storage = k * col_idx_precision *tile*tile
    value_storage = k*tile*tile
    compress_ratio = (offset_storage + col_idx_storage + value_storage) / (M*N)
    print(f'm,n,sparsity,tile:{M,N,1-sparsity,2} offset_storage:{offset_storage/8}B {offset_precision}b col_idx_storage:{col_idx_storage/8}B {col_idx_precision}b compress_storage:{(offset_storage + col_idx_storage + value_storage)/8} Bytes, dense_storage:{M*N/8} Bytes, compress_ratio:{1-compress_ratio}')
```
